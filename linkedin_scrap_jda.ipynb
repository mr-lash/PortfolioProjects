```python
# Import packages

# %pip install selenium

from selenium import webdriver
import time
import pandas as pd
import os

from selenium.webdriver.support.select import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
```

```python
url = 'https://www.linkedin.com/jobs/search/?currentJobId=3167261713&f_TPR=r86400&keywords=junior%20data%20analyst&location=Belgium&refresh=true'
```

```python
driver = webdriver.Edge(executable_path=r"C:/Users/marco/Downloads/edgedriver_win64/msedgedriver.exe")
driver.get(url)
driver.implicitly_wait(10)
```

```python
n=driver.find_elements(By.CLASS_NAME,'results-context-header__job-count')[0].text
```

```python
n
```

```python
y=pd.to_numeric(n)
```

```python
y
```

```python
for _ in range(2, 17):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    try:
        x=driver.find_element(By.XPATH("//button[@aria-label='Cargar m√°s resultados']"))
        driver.execute_script("arguments[0].click();")
        time.sleep(3)

    except Exception:
        time.sleep(3)
```

### Creating the company name column

```python
companyname=[]
```

```python
for j in range(y):
    try:
        company = driver.find_elements(By.CLASS_NAME,'base-search-card__subtitle')[j].text
        companyname.append(company)
    
    except IndexError:
        print("done")
```

```python
companyname
```

### Creating the title name column

```python
titlename=[]
```

```python
for j in range(y):
    try:
        title = driver.find_elements(By.CLASS_NAME,'base-search-card__title')[j].text
        titlename.append(title)
    
    except IndexError:
        print("done")
```

```python
titlename
```

### Creating the company location column

```python
location=[]
```

```python
for j in range(y):
    try:
        where = driver.find_elements(By.CLASS_NAME,'job-search-card__location')[j].text
        location.append(where)
    
    except IndexError:
        print("done")
```

```python
location
```

### Creating the job offer link column

```python
jobList = driver.find_elements(By.CLASS_NAME,'base-card__full-link')
hrefList = []
for e in jobList:
    hrefList.append(e.get_attribute('href'))
```

```python
hrefList
```

### Creating DataFrame's columns

```python
companyfinal=pd.DataFrame(companyname, columns=["company"])
```

```python
titlefinal=pd.DataFrame(titlename, columns=["title"])
```

```python
locationfinal=pd.DataFrame(location, columns=['location'])
```

```python
linkfinal=pd.DataFrame(hrefList, columns=['link'])
```

### Joins to create our table 

```python
final1=companyfinal.join(titlefinal)
```

```python
final1
```

```python
final2=final1.join(locationfinal)
```

```python
final2
```

```python
finaldef=final2.join(linkfinal)
```

```python
finaldef
```

### To CSV file

```python
finaldef.to_csv('juniordataanalystjobs.csv')
```
